## Dirac notation

$\ket{a}$ is the column vector with a $1$ in the entry correcponding to $a \in \Sigma$, with $0$ for all other entries. If $\Sigma = \{0,1\}$:

$\ket{0}=\begin{pmatrix} 1 \\ 0 \end{pmatrix}$

and

$\ket{1}=\begin{pmatrix} 0 \\ 1 \end{pmatrix}$

These are **standard basis vectors**.

Every vector can be expressed uniquely as a linear combination of stadard basis vectors.

$$
\begin{pmatrix}
\frac{3}{4} \\
\frac{1}{4}
\end{pmatrix} = \frac{3}{4}\ket{0} + \frac{1}{4}\ket{1} = \frac{3}{4}\begin{pmatrix} 1 \\ 0 \end{pmatrix} + \frac{1}{4}\begin{pmatrix} 0 \\ 1 \end{pmatrix} = \begin{pmatrix} \frac{3}{4} \\ 0 \end{pmatrix} + \begin{pmatrix} 0 \\ \frac{1}{4} \end{pmatrix} = \begin{pmatrix} \frac{3}{4} \\ \frac{1}{4} \end{pmatrix}
$$
