## Dirac notation

$\ket{a}$ is the column vector with a $1$ in the entry correcponding to $a \in \Sigma$, with $0$ for all other entries. If $\Sigma = \{0,1\}$:

$\ket{0}=\begin{pmatrix} 1 \\ 0 \end{pmatrix}$

and

$\ket{1}=\begin{pmatrix} 0 \\ 1 \end{pmatrix}$

These are **standard basis vectors**.

Every vector can be expressed uniquely as a linear combination of stadard basis vectors.

$$
\begin{pmatrix}
\frac{3}{4} \\
\frac{1}{4}
\end{pmatrix} = \frac{3}{4}\ket{0} + \frac{1}{4}\ket{1} = \frac{3}{4}\begin{pmatrix} 1 \\ 0 \end{pmatrix} + \frac{1}{4}\begin{pmatrix} 0 \\ 1 \end{pmatrix} = \begin{pmatrix} \frac{3}{4} \\ 0 \end{pmatrix} + \begin{pmatrix} 0 \\ \frac{1}{4} \end{pmatrix} = \begin{pmatrix} \frac{3}{4} \\ \frac{1}{4} \end{pmatrix}
$$

## Euclidean norm

For real-valued vectors the Euclidean or $L^2$ norm is:

$$
\|x\|_2=\sqrt{x_1^2 + x_2^2 + \dots + x_n^2}
$$

For complex-valued vectors, with a complex number $z = a + bi$ the magnitude is $|x_i| = \sqrt{a^2 + b^2}$, thus it is:

$$
\|x\|_2=\sqrt{ \sum_{i=1}^{n}{|x_i|^2} }
$$


```python
v = np.array( [2, 3, 6] )
norm = np.linalg.norm(v)
print(norm)
```

```python
v = np.array( [ 4 + 0j, 2 + 2j, 2 - 2j, 2 + 0j ] )
norm = np.linalg.norm(v)
print(norm)
```

## Complex Conjugate and Norm

Let $z = x + iy$ where $x,y\in \mathbb{R}$ and $i^2=-1$.

Then the complex conjugate of $z$ is $\overline{z}=x-iy.$

Compute the product $z\cdot \overline{z}$: $z\overline{z}=(x+iy)(x-iy).$

Expand using distributivity: $(x+iy)(x-iy)=x\cdot x+x\cdot (-iy)+iy\cdot x+iy\cdot (-iy).$

Thus: $=x^2-ixy+ixy-i^2y^2.$

The middle terms cancel: $-ixy+ixy=0.$

And since $i^2=-1$, we get: $-i^2y^2=-(-1)y^2=y^2.$

So the whole expression becomes: $z\overline{z}=x^2+y^2.$
